# AlexNet para classificação do MNIST

Implementação da arquitetura AlexNet adaptada ao conjunto de dados MNIST para treinamento durante 10 épocas e análise dos resultados.

2. Importar bibliotecas

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

from tensorflow.keras import datasets
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D
from tensorflow.keras.layers import Flatten, Dense, Dropout

from sklearn.metrics import confusion_matrix
import seaborn as sns
```

3. Carregar e preparar o MNIST

```python
(x_train, y_train), (x_test, y_test) = datasets.mnist.load_data()

x_train = x_train.reshape(-1, 28, 28, 1).astype("float32") / 255.0
x_test = x_test.reshape(-1, 28, 28, 1).astype("float32") / 255.0

print(x_train.shape)
print(x_test.shape)
```

4. Criar a AlexNet adaptada

```python
model = Sequential()

model.add(Conv2D(96, (3,3), activation='relu', input_shape=(28,28,1)))
model.add(MaxPooling2D(pool_size=(2,2)))

model.add(Conv2D(256, (3,3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2)))

model.add(Conv2D(384, (3,3), activation='relu'))
model.add(Conv2D(384, (3,3), activation='relu'))
model.add(Conv2D(256, (3,3), activation='relu'))

model.add(Flatten())

model.add(Dense(4096, activation='relu'))
model.add(Dropout(0.5))

model.add(Dense(4096, activation='relu'))
model.add(Dropout(0.5))

model.add(Dense(10, activation='softmax'))
```

5. Mostrar arquitetura

```python
model.summary()
```

6. Compilar o modelo

```python
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
```

7. Treinar por 10 épocas

history = model.fit(
    x_train,
    y_train,
    epochs=10,
    validation_data=(x_test, y_test)
)

8. Mostrar Loss e Accuracy

```python
loss, accuracy = model.evaluate(x_test, y_test)

print("Loss:", loss)
print("Accuracy:", accuracy)
```

9. Gráfico de Accuracy

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.title('Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Treino', 'Teste'])

plt.show()
```

10. Gráfico de Loss

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.title('Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Treino', 'Teste'])

plt.show()
```

11. Gerar previsões

```python
y_pred = model.predict(x_test)
y_pred = np.argmax(y_pred, axis=1)
```

12. Matriz de confusão

```python
cm = confusion_matrix(y_test, y_pred)

plt.figure(figsize=(10,8))

sns.heatmap(
    cm,
    annot=True,
    fmt='d',
    cmap='Blues'
)

plt.title('Matriz de Confusão - AlexNet')
plt.xlabel('Predito')
plt.ylabel('Real')

plt.show()
```
